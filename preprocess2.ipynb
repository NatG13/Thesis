```python
#Natalie's code
#1st Feauture group: Engagement (Flow / Routine Play)
#Time Gaps Between Games-> Average threshold between sessions

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Strip spaces and convert timestamp
df['timestamp'] = pd.to_datetime(df_new['timestamp'].str.strip(), format='%d-%b-%y %I.%M.%S.%f %p')
df = df.sort_values(by=['Psn_account', 'timestamp'])

# Calculate time difference between events
df['time_diff'] = df.groupby('Psn_account')['timestamp'].diff().dt.total_seconds() / 60  # in minutes

# Drop NaN values
time_diffs = df['time_diff'].dropna()

# Plot time gap distribution
plt.figure(figsize=(10, 5))
sns.histplot(time_diffs, bins=200, log_scale=(False, True))
plt.xlabel('Time between events (minutes)')
plt.ylabel('Frequency (log scale)')
plt.title('Distribution of time gaps between player events')
plt.grid(True)
plt.show()

# Summary stats to help choose session threshold
print(time_diffs.describe(percentiles=[.5, .75, .9, .95, .99]))
```

```python
#calculate the avg_session_length variable 
#Identify session boundaries using a 10-minute gap
df['new_session'] = (df.groupby('Psn_account')['timestamp'].diff() > pd.Timedelta(minutes=10)) | (df.groupby('Psn_account')['timestamp'].diff().isna())
df['session_id'] = df.groupby('Psn_account')['new_session'].cumsum()

#Calculate session start/end times and duration in minutes
session_durations = df.groupby(['Psn_account', 'session_id'])['timestamp'].agg(['min', 'max'])
session_durations['session_length'] = (session_durations['max'] - session_durations['min']).dt.total_seconds() / 60

#Compute average session length per player
avg_session_length = session_durations.groupby('Psn_account')['session_length'].mean().reset_index()
avg_session_length.rename(columns={'session_length': 'avg_session_length'}, inplace=True)

# Preview result
print(avg_session_length.head())
```

```python
#calculate the session_time_category variable 
# Step 1: Get session start time per session
session_starts = df.groupby(['Psn_account', 'session_id'])['timestamp'].min().reset_index()
session_starts['hour'] = session_starts['timestamp'].dt.hour

#Step 2: Updated 3-category function
def categorize_custom_time(h):
    if 8 <= h < 17:
        return 'Day'
    elif 17 <= h or h < 3:
        return 'Evening'
    else:  # 3 <= h < 8
        return 'Night'

session_starts['session_time_category'] = session_starts['hour'].apply(categorize_custom_time)

# Step 3: Most common time category per player
player_time_category = session_starts.groupby('Psn_account')['session_time_category'] \
                                     .agg(lambda x: x.mode()[0]).reset_index()

# Rename column
player_time_category.rename(columns={'session_time_category': 'preferred_session_time'}, inplace=True)
player_time_category

# # Preview result
# print(player_time_category.head())

# # Step 4: Percentage of players by session time category
# category_counts = player_time_category['preferred_session_time'].value_counts(normalize=True) * 100
# category_counts = category_counts.reset_index()
# category_counts.columns = ['preferred_session_time', 'percentage']

# # Preview
# print(category_counts)

# # Optional: Plot it
# import matplotlib.pyplot as plt

# plt.figure(figsize=(6, 4))
# plt.bar(category_counts['preferred_session_time'], category_counts['percentage'])
# plt.ylabel('Percentage of Players')
# plt.xlabel('Preferred Session Time')
# plt.title('Player Distribution by Preferred Session Time')
# plt.grid(axis='y')
# plt.show()
```

```python
#calculate the total_days_active variable
# Step 1: Let pandas auto-detect the ISO format
df['T_DATE'] = pd.to_datetime(df['T_DATE'])

# Step 2: Count unique active days per player
active_days = df.groupby('Psn_account')['T_DATE'].nunique().reset_index()
active_days.rename(columns={'T_DATE': 'total_days_active'}, inplace=True)

# Preview
print(active_days.head())

# Summary statistics
print(active_days['total_days_active'].describe(percentiles=[.25, .5, .75, .9, .95, .99]))
```

```python
#3rd feature group: Content Immersion (Narrative / Extra Material)
#dlc_count
dlc_cols = ['DLC_1', 'DLC_2', 'DLC_3']

# Step 1: Flag whether each DLC was ever owned by each player
dlc_flags = df.groupby('Psn_account')[dlc_cols].max().reset_index()

# Step 2: Count how many unique DLCs they own (out of 3)
dlc_flags['dlc_count'] = dlc_flags[dlc_cols].sum(axis=1)

# Step 3: Keep only Psn_account and dlc_count
dlc_count = dlc_flags[['Psn_account', 'dlc_count']]
dlc_count

# # Step 4: Redo distribution
# dlc_distribution = dlc_count['dlc_count'].value_counts(normalize=True).sort_index() * 100
# dlc_distribution = dlc_distribution.reset_index()
# dlc_distribution.columns = ['dlc_count', 'percentage']

# print("Corrected DLC Ownership Distribution (%):")
# print(dlc_distribution)
# dlc_count
```

```python
#mode_diversity

# Step 1: Load only what's needed
original_modes = pd.read_csv('data.csv', usecols=['Psn_account', 'Mission_mode'])

# Step 2: Clean mission mode names
original_modes['Mission_mode_clean'] = df['Mission_mode'].astype(str).str.strip().str.title()
original_modes['Mission_mode_clean'] 

# Step 3: Count distinct modes per player
mode_diversity = original_modes.groupby('Psn_account')['Mission_mode_clean'].nunique().reset_index()
mode_diversity.rename(columns={'Mission_mode_clean': 'mode_diversity'}, inplace=True)
mode_diversity

# # Step 4: Check the distribution
# print(mode_diversity['mode_diversity'].value_counts())

# # Get distribution of mode_diversity values
# diversity_distribution = mode_diversity['mode_diversity'].value_counts(normalize=True).sort_index() * 100

# # Format as DataFrame
# diversity_percentages = diversity_distribution.reset_index()
# diversity_percentages.columns = ['mode_diversity', 'percentage']

# # Display
# print("Percentage of Players by Mode Diversity:")
# print(diversity_percentages)
```

```python
# Step 2: Get peak rank per player
peak_rank = peak_rank.groupby('Psn_account')['peak_rank'].max().reset_index()

# Step 3: Categorize peak rank only
def categorize_peak(rank):
    if pd.isna(rank):
        return 'Unknown'
    elif 0 <= rank <= 15:
        return 'Beginner'
    elif 16 <= rank <= 30:
        return 'Advanced'
    elif 31 <= rank <= 42:
        return 'Hardcore'
    elif 43 <= rank <= 44:
        return 'Prestige'
    elif rank == 45:
        return 'Final'
    else:
        return 'Unknown'

peak_rank['peak_rank_category'] = peak_rank['peak_rank'].apply(categorize_peak)

# Preview
peak_rank.head()
```

```python
#combine datframes 
#Merging into one new dataset

# Merge all dataframes step-by-step on Psn_account
from functools import reduce

# List of all feature DataFrames
feature_dfs = [
    avg_session_length,
    player_time_category,      # preferred_session_time
    active_days,               # total_days_active
    peak_rank, 
    avg_score,
    dlc_count,
    mode_diversity
]

# Reduce-merge all on Psn_account
from functools import reduce
all_features = reduce(lambda left, right: pd.merge(left, right, on='Psn_account', how='inner'), feature_dfs)

# Preview the combined dataset
print("âœ… Final Combined Feature Table:")
all_features.head()
```

```python
# One-hot encode preferred_session_time while keeping original column
time_dummies = pd.get_dummies(all_features['preferred_session_time'], prefix='session_time')

# Concatenate the dummies WITHOUT dropping the original column
all_features_encoded = pd.concat([all_features, time_dummies], axis=1)

# Preview result
print("âœ… Feature Table with Original Column + One-Hot Encoding:")
all_features_encoded.head()
```
